```
%scala

val movies_df = spark.read.option("header",true).option("inferSchema",true).csv("/mnt/rawzone/internal/aic_blitz_decrypt/movies.csv")
movies_df.write.mode("overwrite").parquet("/mnt/rawzone/internal/aic_blitz_decrypt/movies_parq.parquet")
```

```
movies_df.write.mode("overwrite").format("parquet").saveAsTable("/mnt/rawzone/internal/aic_blitz_decrypt/movies_parquet")
```

JSON to DELTA

```
%scala
val jsonDf = spark.read.option("multiline", "true").json("/mnt/rawzone/internal/aic_blitz_decrypt/*.json")
display(jsonDf)
```

Create Delta against JSON

```
%scala
jsonDf.write.mode("overwrite").format("delta").saveAsTable("Meyya_jsonDataTable")
```
